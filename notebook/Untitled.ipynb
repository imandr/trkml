```python
import math

def rename_hits(hits, truth):
    hits_truth = hits.set_index("hit_id").join(
            truth.set_index("hit_id"), 
            lsuffix="hit_", 
            rsuffix="truth_")
    #print hits_truth_filtered
    #hits_truth_filtered["xabs"] = hits_truth_filtered.apply(lambda row: abs(row["x"]), axis=1)
    #hits_truth_filtered["yabs"] = hits_truth_filtered.apply(lambda row: abs(row["y"]), axis=1)
    #hits_truth_filtered["zabs"] = hits_truth_filtered.apply(lambda row: abs(row["z"]), axis=1)

    hits_truth["r2"] = np.square(hits_truth["x"]) + np.square(hits_truth["y"])
    hits_truth["zabs"] = np.abs(hits_truth["z"])
    
    hits_sorted = hits_truth.sort_values(["zabs", "r2", "x", "y"]).copy()
    #print hits_sorted.head()
    particles = {}
    inx = 1
    ipids = []
    for i, row in hits_sorted.iterrows():
        pid = long(row["particle_id"])
        if pid == 0:
            ipid = 0
        else:
            ipid = particles.get(pid)
            if ipid is None:
                ipid = inx
                inx += 1
                particles[pid] = ipid
        ipids.append(ipid)

    out = hits_sorted.drop("particle_id", axis=1)
    out["ipid"] = np.array(ipids, dtype=np.int16)

    
    return out

renamed = rename_hits(hits, truth)  
```
